```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pix2num(i,j):
    return(30*i+j)

def num2pix(n):
    i = n//30
    j = n%30
    return i,j

def sampleSlice(i,mu,cov,x):
    mean = mu[i]
    var = cov[0][0]
    x[i] = mu[i] + np.random.standard_normal()*np.sqrt(var)
    return x

def createImage(x):
    img = np.zeros((30,30))
    for i in range(30):
        for j in range(30):
            img[i,j] = x[pix2num(i,j)]
    return img
```

```python
mu = np.zeros(900)
for i in range(30):
    for j in range(30):
        mu[pix2num(i,j)] = (i+j)/100
```

```python
plt.imshow(createImage(mu))
```

```python
cov = np.zeros((900,900))
for i in range(900):
    cov[i,i] = 0.1
```

```python
imgs = []
sample = 0.5*np.ones(900)
for t in range(40):
    for i in range(900):
        sample = sampleSlice(i,mu,cov,sample)
    if (t+1)%10 == 0:
        img = createImage(sample)
        imgs.append(img)
```

```python
len(imgs)
```

```python
fig, axs = plt.subplots(1,5)
axs[0].imshow(createImage(mu))
plt.axis('off')
for i in range(4):
    axs[i+1].imshow(imgs[i])
plt.show()
```
