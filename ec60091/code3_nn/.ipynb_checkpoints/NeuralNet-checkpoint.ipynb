```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = 'Snails.csv'

df = pd.read_csv(train_file)
label_dict = {'M':0, 'I': 1, 'F': 2}
idx_to_label = {0:'Male', 1: 'Both', 2:'Female'}
features_idxs = {feat:idx for idx, feat in enumerate(df.columns[1:-1])}

def get_label(x):
    return label_dict[x]

def one_hot(x):
    vector = [0]*3
    vector[int(x)] = 1
    return vector

### scatter plot using two features for the clusters
def scatter_plot(data, labels, feature_1 = 'length', feature_2 = 'rings'):
    plt.style.use('default')
    idx1, idx2 = features_idxs[feature_1], features_idxs[feature_2]
    
    color_dict = {}
    for d,l in zip(data, labels):
        label = idx_to_label[np.argmax(l)]
        if label not in color_dict: color_dict[label] = []
        color_dict[label].append(d)
        
    for k,v in color_dict.items():
        plt.scatter(np.array(v)[:,idx1],np.array(v)[:,idx2], label = k)
        
    plt.xlabel(feature_1)
    plt.ylabel(feature_2)
    plt.legend()
    plt.show()
    
df['sex'] = df['sex'].apply(get_label)

data = np.array(df.to_numpy()[:,1:-1], dtype = 'float')
labels = np.array([one_hot(x) for x in df.to_numpy()[:,-1]], dtype = 'float')

random_idxs = np.random.permutation(np.arange(data.shape[0]))
train_idxs, val_idxs = random_idxs[:int(0.8*(len(random_idxs)))], random_idxs[int(0.8*(len(random_idxs))):]

train_data, train_labels = data[train_idxs], labels[train_idxs]
val_data, val_labels = data[val_idxs], labels[val_idxs]

mean, std = train_data.mean(axis = 0), train_data.std(axis = 0)

plt.title('Scatter Plot with two features before Z-normalisation')
scatter_plot(train_data, train_labels)

train_data = (train_data-mean)/std
val_data = (val_data-mean)/std

plt.title('Scatter Plot with two features after Z-normalisation')
scatter_plot(train_data, train_labels)
```

```python
class NeuralNet(object):
    def __init__(self, input_size, hidden_size, output_size, std=1e-4):
        self.params = {}
        ## initialise matrices of the neural network
        self.params['W1'] = std * np.random.randn(input_size, hidden_size)
        self.params['b1'] = np.zeros(hidden_size)
        self.params['W2'] = std * np.random.randn(hidden_size, output_size)
        self.params['b2'] = np.zeros(output_size)

    def loss(self, X, y=None):
        # Get variables from the params dictionary
        W1, b1 = self.params['W1'], self.params['b1']
        W2, b2 = self.params['W2'], self.params['b2']
        N, D = X.shape
        
        # Compute the forward pass
        X1 = X.dot(W1) + b1
        X1_act = 1/(1+np.exp(-X1))
        scores = X1_act.dot(W2) + b2
        scores_act = 1/(1+np.exp(-scores))

        # Without labels, no loss needed
        if y is None:
            return scores_act

        # Compute the loss
        loss = 0.5*np.sum(np.power(scores_act-y, 2))/N

        # Backward pass: compute gradients
        dScores_act = (scores_act-y)/N
        dScores = scores_act * (1-scores_act) * dScores_act
#         print(dScores.shape, scores.shape)
        dW2 = X1.T.dot(dScores) #/ N
        db2 = np.ones((1,N)).dot(dScores) #/ N

        dX1_act = (dScores).dot(W2.T)
        dX1 = dX1_act*X1_act*(1-X1_act)
#         print(dX1.shape, X1.shape)
        dW1 = X.T.dot(dX1) #/ N
        db1 = np.ones((1,N)).dot(dX1) #/ N
#         print(dW1.shape, W1.shape)
        grads = {'W1':dW1, 'b1':db1, 'W2':dW2, 'b2':db2}
#         input()
        return loss, grads

    def train(self, X, y, X_val, y_val, batch_size = 1,
              learning_rate=1e0, num_iters=500):
        num_train = X.shape[0]
        # Record history of various metrics
        loss_history = []
        train_acc_history = []
        val_acc_history = []

        for it in range(num_iters+1):
            idxs = np.random.choice(num_train, batch_size)
#             idxs = np.random.permutation(np.arange(batch_size))
            X_batch = X[idxs]
            y_batch = y[idxs]

            # Compute loss and gradients using the current minibatch
            loss, grads = self.loss(X_batch, y=y_batch)
            self.params['W1'] -= learning_rate*grads['W1']
            self.params['W2'] -= learning_rate*grads['W2']
            self.params['b1'] -= learning_rate*(np.squeeze(grads['b1']))
            self.params['b2'] -= learning_rate*(np.squeeze(grads['b2']))


            # Check accuracy
            train_acc = (self.predict(X_batch) == np.argmax(y_batch, axis=1)).mean()
            val_acc = (self.predict(X_val) == np.argmax(y_val, axis=1)).mean()
            
            if it % 100 == 0:
                print('iteration %d / %d: loss %f val_acc %f' % (it, num_iters, loss, val_acc))
            
            #store history
            train_acc_history.append(train_acc)
            val_acc_history.append(val_acc)
            loss_history.append(loss)

        results_dict = {
          'loss_history': loss_history,
          'train_acc_history': train_acc_history,
          'val_acc_history': val_acc_history,
        }
        
        return results_dict

    def predict(self, X):
        scores = self.loss(X)
        y_pred = np.argmax(scores, axis=1)
        return y_pred
```

```python
input_size, hidden_size, output_size = train_data.shape[1], 8, 3
nn = NeuralNet(input_size, hidden_size, output_size)
results_dict = \
nn.train(train_data, train_labels, val_data, val_labels, batch_size = train_data.shape[0], num_iters = 500)

plt.plot(results_dict['loss_history'])
plt.ylabel('Cost (Mean Square Error)')
plt.xlabel('Number of epochs')
plt.title('Variation of training loss')
plt.show()

plt.plot(results_dict['train_acc_history'], label = 'Train Acc')
plt.plot(results_dict['val_acc_history'], label = 'Validation Acc')
plt.ylabel('Accuracy')
plt.xlabel('Number of epochs')
plt.title('Variation of accuracies')
plt.legend()
plt.show()

test_data = np.array([[100,50,20,55.5,42,23,35,11], [110,74,25,153.6,47.4,15.5,11,10],
                      [106,73,16,70.3,47.4,29.9,33,19], [94,81,20,132.9,33.5,34.2,38,10]])
test_data = (test_data - mean)/std

test_labels = nn.predict(test_data)

for idx, label in enumerate(test_labels):
    print('For test-data point {}, label = {}'.format(idx, idx_to_label[label]))
```
